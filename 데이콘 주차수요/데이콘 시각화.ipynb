```python
import pandas as pd
import sweetviz
```

```python
train = pd.read_csv('./new_train.csv', encoding = 'cp949')
test = pd.read_csv('./new_test.csv', encoding = 'cp949')
```

```python
dacon_report2 = sweetviz.compare([train, "Train"],[test, "Test"],"등록차량수")
```

```python
dacon_report2.show_html(".\dacon_report2.html")
```

```python
import pandas_profiling
from pandas_profiling import ProfileReport

#profile 파일 만들기
profile = ProfileReport(train,title="train data set")

#html 파일로 꺼내기
profile.to_file(output_file = "train_profile.html")
```

```python
#profile 파일 만들기
profile2 = ProfileReport(test,title="test data set")

#html 파일로 꺼내기
profile2.to_file(output_file = "test_profile.html")
```
